# Analysing CSV files
Begin by initialising appropriate packages and dowloading CSV file, which in this case would be COVID-19 data from some API

```julia
using Downloads, DataFrames, CSV, Plots, Dates

# Use command Downloads.download() to download file from the internet 
# First argument is to input the link of the file for download
# Second argument in the command is the destination name of the file on local machine

# Downloads.download() always take two arguments!

download_covid = Downloads.download("https://api.covidtracking.com/v1/us/daily.csv",
                                    "covid_us.csv")
```

```julia
# Reading the specific columns from the downloaded CSV file to a DataFrame

columns = [:date, :totalTestResultsIncrease]
fmt = "yyyymmdd"
t = Dict(:date => Date) # Declaring the dictionary to map ':date' column as 'Date' object


# Declaring DataFrame for importing the data from CSV file

covid_df = CSV.read("covid_us.csv",
                    DataFrame,
                    dateformat = fmt, # Specifies the formatting used for date (from earlier)
                    select = columns, # Specifies which columns (from earlier) to read
                    types = t) # Specifies any mapping

# Rearranges the Data Frame in such a way that to filter for 
# 'total.TestResultsIncrease' larger than 0
covid_df = sort(filter(row -> row.totalTestResultsIncrease > 0, covid_df))

# Show output of the last 5 rows of the filtered and sorted DataFrame
last(covid_df, 5)
```

```julia
# Auxilliary functions to determine rows of the DataFrame

nrows = size(covid_df,1)
println("Number of rows: $nrows")
ncolumns = size(covid_df, 2)
println("Number of columns: $ncolumns")
df_size = size(covid_df)
```

```julia
# To plot the data in form using RollingFunction library

using RollingFunctions

# plot daily test increase as sticks
Plots.plot(covid_df.date,
    covid_df.totalTestResultsIncrease,
    seriestype=:sticks,
    label="Test Increase",
    title = "USA Total Testing Capacity",
    lw = 2)

# 7-day average using rolling mean
window = 7
average = rollmean(covid_df.totalTestResultsIncrease, window)

# we mutate the existing plot
Plots.plot!(covid_df.date,
    cat(zeros(window - 1), average, dims=1),
    label="7-day Average",
    lw=3)
```
