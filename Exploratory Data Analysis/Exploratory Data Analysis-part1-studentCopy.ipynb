# Exploratory Data Analysis

## Introduction
<br>
<div style="text-align: justify">In the previous chapter, we have seen that data preprocessing is an important step in the data science pipeline. Once we get the preprocessed data, we have to choose suitable machine learning algorithms to model the data. However, before applying machine learning, we have to answer the following questions:</div>

- How to find the structure of the data?
- How to test assumptions about the data?
- How to select the features that can be used for machine learning methods?
- How to choose suitable machine learning algorithms to model our dataset?

<div style="text-align: justify">Exploratory Data Analysis (EDA) is a process to get familiar with the structure and important features of a dataset. EDA helps us answer the aforementioned questions by providing us with a good understanding of what the data contains. EDA explores the preprocessed data using suitable visualization tools to find the structure of the data, its salient features, and important patterns.</div>

<div style="text-align: justify">EDA is also intended to define and refine the selection of
important features. Once EDA is complete, we can perform
more complex modeling and machine learning tasks on
the selected features such as clustering, regression, and
classification.</div>

<div style="text-align: justify">The goal of EDA is to make sure that the dataset is ready to be
deployed in a machine learning algorithm. EDA is a valuable
step in a data science project, validating the results, and making
the interpretation of the results simpler in the desired context.
In the ensuing sections, we explain the process of exploratory
data analysis along with practical Python examples.</div>

## Revealing Structure of Data

<br>

<div style="text-align: justify">Knowing the underlying structure of our data enables us to
use appropriate machine learning methods for data modeling
and future predictions using these models. In EDA, several
techniques are employed to reveal the structure of the data.
These include:</div>

<div style="text-align: justify"><b>Univariate visualization</b> is used to generate summary statistics
for each feature or variable in a dataset. We summarize our
dataset through descriptive statistics that uses a variety of
statistical measurements to better understand the dataset.
    It is also known as <b>data profiling</b>. The goal of univariate
visualization is to have a solid understanding of the data so
we can start querying and visualizing our data in various ways.
It uses visualization tools such as bar plots and histograms to
reveal the structure of the data.</div>

<div style="text-align: justify"><b>Bivariate visualization</b> is carried out to find the relationship
between two variables in a given dataset, where one of the
two variables can be the target variable of interest. It uses correlations, scatter, and line plots to reveal the structure of
the data.</div>

<div style="text-align: justify"><b>Multivariate visualization</b> is employed to understand
interactions between different fields in the dataset. It uses
line plots, scatter plots, and matrices with multiple colors to
understand the relationship between various features of a
dataset.</div>
<br>


<div style="text-align: justify">Revealing the underlying structure of data enables us to
discover patterns, spot anomalies such as missing values and
outliers, and check assumptions about the data.</div>

## Plots and Charts

### Line Plot
<br>
<div style="text-align: justify">Line plots are predominantly useful for conveying changes
over space or time. Line plots are mostly used to plot data
along a scale divided into equal intervals, for example, time.
Let us generate a simple line plot.</div>

```python
# Simple x and y coordinates
x1 = [1,2,3]
y1 = [2,4,1]

# Plot the line chart
plt.plot(x1,y1, label = 'line 1 ')
plt.show()
```

<div style="text-align: justify">In this graph, we generate three values of variable x1 and y1. To
generate a line plot via the pyplot module, we call the function
plot(), pass it the values for the x and y axes. It is important to
mention that we are using plt as an alias for pyplot.</div>

<div style="text-align: justify">We can add titles, labels, and legends to the generated plots.
To add titles, labels, and legend to a plot, we use the title,
xlabel, ylabel, and legend methods of the pyplot module, respectively. We pass string values to these methods, which
appear on the plots, as shown next.</div>

```python
# Simple x and y coordinates
x1 = [1,2,3]
y1 = [2,4,1]
 
# Plot the line chart
plt.plot(x1,y1, label = 'line 1 ')

# naming the x axis
plt.xlabel('x-axis')

# naming the y axis
plt.ylabel('y-axis')

# giving a title to my graph
plt.title('My line chart')
plt.legend()
plt.show()
```

<div style="text-align: justify">Multiple plots can be drawn on the same figure, as shown in
the Python script given below.</div>

```python
# Generating x and y points to plot
x1 = np.arange(0, 10*(np.pi),0.1)
y1 = np.cos(x1) # cosine function from Numpy library

# potting the points
plt.plot(x1, y1, label = 'cosine')
#plt.legend()

# shows the legend with labels
# naming the x axis
plt.xlabel('Angle')

# naming the y axis
plt.ylabel('Amplitude')
#plt.show()

# Generating x and y points for the second plot
# decaying cosine function
x2 = np.arange(0, 10*(np.pi), 0.1)
y2 = np.cos(x2)*0.1*np.arange(10*(np.pi), 0, -0.1)

# potting the points
plt.plot(x2,y2, label = 'decaying cosine')
plt.legend()
plt.title('Two Functions on the same graph')

# show a legend on the plot
# gives a title to the plot

# shows the graph, and removes the text output.
```

<div style="text-align: justify">Let us plot <b>Avg. Area Income</b> against <b>Price</b> from US_Housing

### Scatter Plot
<br>
<div style="text-align: justify">A scatter plot is used to visualize any two variables from a
given dataset in two dimensions. It uses dots or marks to plot
values of two variables, one along the x-axis and the other
along the y-axis.</div>

<div style="text-align: justify">Scatter plots allow us to observe the relationship between
the variables. If an increase in one variable causes an increase
in another variable and vice versa, we can infer there is a
positive linear relationship between two variables. However, if
increasing the first variable reveals a decrease in the second
variable, we say that there is a negative linear relationship
between both variables. For example, let us generate a scatter
    plot between <b>Avg. Area Income</b> and <b>Price.</b></div>

```python
# Adjusting the figure size using rcParams function

plt.rcParams['figure.figsize'] = [6,4]



# Scatter plot
plt.scatter(housing_price['Avg. Area Income'], housing_price['Price'], color = 'red', marker = 'o')


# Giving title and label names
plt.title('Avg. Area Income vs Price', fontsize = 14)
plt.xlabel('Avg. Area Income', fontsize = 14)
plt.ylabel('Price', fontsize = 14)
plt.grid(True)

plt.show()
```

<div style="text-align: justify">This plot shows that there is a positive linear relationship
between variables Avg. Area Income and Price. To increase
the size of plot for better readability, we have used rcParams
option of pyplot module.</div>

<br>
plt.rcParams[‘figure.figsize’] = [12,8]

<div style="text-align: justify">If the points or dots are color-coded in a scatter plot, additional
variables can be shown in two-dimensions. For example,
    let us create the <b>Avg. Area Income</b> against <b>Price</b> and <b>Area
    Population</b> against <b>Price</b> by color-coding the plots on the
same figure.</div>

```python
#Exercise: Plot Area Population Vs Price - use scatter, blue and x marker


# Adjusting the figure size using rcParams function
plt.rcParams['figure.figsize'] = [12,8]

# Scatter plot
plt.scatter(housing_price['Area Population'], housing_price['Price'], color = 'red', marker = 'o', label = 'Area Population vs Price')
plt.scatter(housing_price['Avg. Area Income'], housing_price['Price'], color = 'blue', marker = 'o', label = 'Avg. Area Income vs Price')

# Giving title and label names
plt.title('Avg. Area Income / Area Population vs Price', fontsize = 14)
plt.xlabel('Avg. Area Income (Blue) , Area Population (Red) ', fontsize = 14)
plt.ylabel('Price', fontsize = 14)
plt.grid(True)
plt.legend()

plt.show()

```
